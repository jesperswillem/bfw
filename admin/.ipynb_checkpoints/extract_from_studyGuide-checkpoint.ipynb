```python
import urllib.request
import markdownify
from pyhtml2pdf import converter
import os
from bs4 import BeautifulSoup


def read_html(htmlpage):
    fp = urllib.request.urlopen(htmlpage)
    mybytes = fp.read()

    mystr = mybytes.decode("utf8")
    fp.close()

    return mystr

data = read_html("https://studiegids.universiteitleiden.nl/en/studies/8969/bio-pharmaceutical-sciences#tab-1")
data = data.split()

# retrieve course html links
#courses = {}
for line in data:
    if 'href' in line and 'courses' in line:
        link = line.split('>')[0].strip('href=').strip('"')
        resp = read_html(link)

        # trying to get course data
        page = BeautifulSoup(resp, "html.parser")

        dt = []
        dd = []
        for a in page.find_all('dt'):
            a = str(a)
            a = a.split('>')[1]
            a = a.split('<')[0]
            dt.append(a)

        for a in page.find_all('dd'):
            a = str(a)

            # this logic at the moment returns empty teacher fields
            a = a.split('>')[1]
            a = a.split('<')[0]               
            dd.append(a)

        course_data = dict(zip(dt, dd))
        if not 'Level' in course_data or not 'Catalog number' in course_data:
            if 'Studiegidsnummer'in course_data:
                course_data['Catalog number'] = course_data['Studiegidsnummer']

                if not 'Level' in course_data:
                    for line in str(page).split():
                        if 'href' in line and 'courses' in line:
                            print(line)
                            print(course_data)

                    break                  

            else:    
                for line in str(page).split():
                    if 'href' in line and 'courses' in line:
                        print(line)
                        print(course_data)
                continue
        # create folder logic
        if not os.path.isdir(course_data['Level']):
            os.mkdir(course_data['Level'])

        # create folder logic]
        if not os.path.isdir('{}/{}'.format(course_data['Level'],course_data['Catalog number'])):
            os.mkdir('{}/{}'.format(course_data['Level'],course_data['Catalog number']))
        
        # save pdf and markdown of course information
        converter.convert(link, '{}/{}/info.pdf'.format(course_data['Level'],course_data['Catalog number']))
        #print(course_data)
        #break
        
        # markdown broken at the moment
        #h = markdownify.markdownify(page, heading_style="ATX")         
        #with open('{}/{}/info.md'.format(course_data['Level'],course_data['Catalog number']), 'w') as outfile:
        #    outfile.write(h)
```
